# missing .. handled case 2,3,4,5,33,44

```python
miss_mvar2=train_df.mvar2.dropna().mode()[0]
miss_mvar3=train_df.mvar3.dropna().mode()[0]
miss_mvar4=train_df.mvar4.dropna().mode()[0]
miss_mvar5=train_df.mvar5.dropna().mode()[0]
miss_mvar33=train_df.mvar33.dropna().mode()[0]
miss_mvar44=train_df.mvar44.dropna().mode()[0]
```

```python
train_df['mvar2']=train_df['mvar2'].fillna(miss_mvar2)
train_df['mvar3']=train_df['mvar3'].fillna(miss_mvar3)
train_df['mvar4']=train_df['mvar4'].fillna(miss_mvar4)
train_df['mvar5']=train_df['mvar5'].fillna(miss_mvar5)
train_df['mvar33']=train_df['mvar33'].fillna(miss_mvar33)
train_df['mvar44']=train_df['mvar44'].fillna(miss_mvar44)
```
