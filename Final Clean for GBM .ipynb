# Training clean after merging

```python
train_df=pd.read_csv('Training_Missing_removed.csv',index_col='application_key')
colm= [ i for i in range(48)]
train_df.columns=colm
med_colm=[0,1,2,3,4,8,9,18,19,27,28,32,33,35,36,37,38,41,42,43]
zero_replace_colm=[5,6,7,10,11,12,14,15,16,17,20,21,22,23,24,25,26,
                  29,30,31,34,39,40,44,45]
train_df[med_colm]=train_df[med_colm].replace(np.NaN,train_df[med_colm].median())
train_df[zero_replace_colm]=train_df[zero_replace_colm].replace(np.NaN,0)
train_df[[6]]=train_df[[6]].values+train_df[[7]].values
train_df[[15]]=train_df[[16]].values+train_df[[17]].values+train_df[[15]].values
train_df[[21]]=train_df[[21]].values+train_df[[22]].values
train_df[[25]]=train_df[[25]].values+train_df[[26]].values
train_df[[29]]=train_df[[29]].values+train_df[[30]].values
train_df[[39]]=train_df[[39]].values+train_df[[40]].values
train_df[[44]]=train_df[[44]].values+train_df[[45]].values
drop_colm=[7,16,17,22,26,30,40,45]
replace1_colm=[6,15,21,25,29,39,44]
train_df=train_df.drop(drop_colm,axis=1)
train_df[train_df[replace1_colm]>1]=1
# train_df.to_csv('Training_cleaned_merge3oct.csv')
# train_df.shape
```
